## 【問題4】欠損値への対応

欠損値に関して以下のことを行い、レポートしてください。

* 各特徴量の欠損値の有無を確認する。

欠損値の確認にはmissingnoというライブラリを活用しましょう。このライブラリにより欠損値の有無が可視化できます。インストールや使い方に関しては以下のページをご覧ください。

[ResidentMario/missingno: Missing data visualization module for Python.](https://github.com/ResidentMario/missingno)

* 欠損値が何パーセント含まれるかを確認する。
* 5つ以上欠損値がある特徴量（列）は削除する。
* 欠損値があるサンプル（行）は削除する。

```python
import missingno as msno
%matplotlib inline

# 各特徴量の欠損値の有無を確認する

msno.matrix(house_variables)
```

```python
from decimal import Decimal, ROUND_HALF_UP

# 欠損値が何パーセント含まれるかを確認する。

# isnaメソッドを使い、全ての値をTrue: Null, False: 値有り、に置換する。
hf_boolean = house_variables.isna()

# sumメソッドにてTrue: 1を合算して、欠損数を算出する。
hf_missing_num = hf_boolean.sum()

# 欠損数をlen関数で割ってパーセント数を算出する。
hf_missing_ratio = (hf_missing_num / len(house_variables)).round(4) * 100
# hf_missing_ratio

hf_missing_dataframe = pd.concat([hf_missing_num, hf_missing_ratio], axis=1, ignore_index=False)
hf_final_dataframe = hf_missing_dataframe.rename(columns={0: 'num', 1: 'missing_ratio'})
hf_final_dataframe
```

```python
# 5つ以上欠損値がある特徴量（列）は削除する。
# 1つの列に欠損値が5以下の場合、欠損値ではない要素が1455個（最大1460個）のため、thresh=1455とする。

df_null_less_than_five = house_variables.dropna(thresh=1455, axis=1)
df_null_less_than_five
```

```python
# 欠損値があるサンプル（行）は削除する。
# 欠損値が1つでもあれば、行ごと削除する。

df_no_null = house_variables.dropna()
df_no_null
```

```python
# df_null_less_than_fiveを用いる場合。

df_no_null = df_null_less_than_five.dropna()
df_no_null
```

## 【問題5】用語の調査

以下の２つの単語を知る必要があります。これらを調査してまとめてください。

* 尖度
* 歪度

## 【解答】

* 尖度：分布の尖り具合を示すもの。尖度が大きい程、尖った分布をしている。正規分布の場合、尖度 = 3となり、これが基準となる。
* 歪度：分布が正規分布からどれだけ歪んでいるかを表す統計量。左右対称を示す指標であるため、正規分布の場合は0, 左に偏った分布の場合は正の値、右に偏った場合は負の値となる。

## 【問題6】分布の確認

* seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
* 目的変数に対して対数変換を行う。
* 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。

```python
from matplotlib import pyplot as plt
from scipy.stats import norm
import seaborn as sns
%matplotlib inline

# 目的変数：SalePriceを用い、分布を表示。
sns.distplot(
    house_variables['SalePrice'], 
    label='SalePrice',
    kde_kws={'label': 'kde', 'color':'k'},
    fit=norm,
    fit_kws={'label': 'norm', 'color':'red'},
    rug=False
)

plt.legend()
plt.show()

# 歪度と尖度を算出する。

SalePrice_skewness = house_variables['SalePrice'].skew()
SalePrice_kurtosis = house_variables['SalePrice'].kurt()

print("歪度（Skewness）：{}".format(SalePrice_skewness))
print("尖度（Kurtosis）：{}".format(SalePrice_kurtosis))
```

```python
import numpy as np

# 目的変数に対して対数変換を行う
# numpyのnp.log(a)を用いる。

log_house_variables = np.log(house_variables['SalePrice'])

# 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

sns.distplot(
    log_house_variables,
    label='SalePrice',
    kde_kws={'label': 'kde', 'color':'k'},
    fit=norm,
    fit_kws={'label': 'norm', 'color':'red'},
    rug=False
)

plt.legend()
plt.show()

# 歪度と尖度を算出する。

SalePrice_skewness = house_variables['SalePrice'].skew()
SalePrice_kurtosis = house_variables['SalePrice'].kurt()

print("歪度（Skewness）：{}".format(SalePrice_skewness))
print("尖度（Kurtosis）：{}".format(SalePrice_kurtosis))
```

## 【考察】

もとのデータが正規分布しない場合、対数変換を用いて、正規分布化させる。
それによって、平均値など、統計的分析が容易になる。
よって、正規分布に近づければ、機械学習上、良質な結果がでる。

## 【問題7】相関係数の確認

特徴量の確認をしていきます。それぞれに対してマークダウンで説明や考察を書いてください。

* 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。
* 目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。
* 選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。
* 選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。

```python
# 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。

from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

# データセットの母集団から全ての要素"house_variables"をcorr()メソッドを用い、相関行列化する。
df_corr = house_variables.corr()

# df_corrを用い、ヒートマップ化する。

plt.figure(figsize=(15, 12)) 
df_corr_heatmap = sns.heatmap(df_corr, vmax= 1.0, vmin= -1.0, annot=False)
```

## 【考察１】
全体的なヒートマップを描写したものの、一概に関係性を示した分布が明瞭ではない。

目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。

相関の高い（と思われる）特徴量10個
1. YearBuilt:建築日
2. YearRemodAdd：改築日
3. 1stFlrSF：1階の面積（ft^2)
4. GrLivArea：地上階層の総面積（ft^2）
5. BsmtFullBath：地下階層のフルサイズのトイレ・入浴室の有無
6. TotRmsAbvGrd：トイレ・入浴室を除いた、地上階層の総面積（ft^2）
7. GarageYrBlt：車庫の建築日
8. GarageCars：車庫のサイズ（車何台分か）
9. GarageArea：車庫の面積（ft^2）
10. OpenPorchSF：屋外型ポーチの面積（ft^2）


```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline

selected_df = house_variables[['YearBuilt', 'YearRemodAdd', '1stFlrSF', 'GrLivArea', 'BsmtFullBath', 'TotRmsAbvGrd', 'GarageYrBlt', 'GarageCars', 'GarageArea', 'OpenPorchSF']]

# データセットから選択した10要素のみの"selected_df"をcorr()メソッドを用い、相関行列化する。
df_corr = selected_df.corr()

# df_corrを用い、ヒートマップ化する。
plt.figure(figsize=(12, 9)) 
sns.heatmap(df_corr, vmax= 1.0, vmin= 0, annot=True)
```

## 【考察２】

選んだ10個の特徴から、特に、お互いの相関係数が高い組み合わせを3つ選ぶとする。。

* "YearBuilt" と "GarageYrBlt" は、それぞれ、建物と車庫の建設日のため、高い相関性があることは推測できる。
* "YearBuilt" と "TotRmsAbvGrd" も 建物の建設日と地上階層の総面積は相関性が高いことも伺える。
* "GarageYrBlt" と "GarageCars" と "GarageArea" も同じ車庫に関する内容のため、お互いの相関性も高いと推測できる。
