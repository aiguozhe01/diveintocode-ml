# Week4授業前課題2 住宅価格の予測

## 【問題1】練習のために特徴量の選択

1. csvファイルから説明変数（特徴量）GrLivAreaとYearBuiltの抽出
2. 目的変数SalePriceを抜き出す。

```python
import pandas as pd
import numpy as np

# pandasを用いて、csvファイルをデータフレームとして読み込む。
house_variables = pd.read_csv("house_prices_train.csv")

# 変数に代入したのち、説明変数（X_house_var）と目的変数（y_house_var）を抽出する。
# dfをnumpy.array化
X_house_var = house_variables[['GrLivArea', 'YearBuilt']].values
y_house_var = house_variables[['SalePrice']].values

print(house_variables[['GrLivArea', 'YearBuilt']])
```

## 【問題2】線形回帰による推定と評価

1. データを簡単に分析、前処理を行う。
2. 線形回帰による学習（fit）を行う。
3. 推定（predict）を行う。
4. 評価（MSE）を行う。
5. 可視化（散布図）する。

```python
# データを簡単に分析、前処理を行う。
# GrLiveAreaとSalePriceの関係を分析

import matplotlib.pyplot as plt
%matplotlib inline


plt.figure(figsize=(12, 8))
plt.scatter(house_variables['GrLivArea'], house_variables['SalePrice'])   # 平均部屋数と住宅価格の散布図をプロット

plt.title('Scatter Plot of GrLivArea vs SalePrice')   # 図のタイトル
plt.xlabel('Above grade (ground) living area square feet [GrLivArea]')   # x軸のラベル
plt.ylabel('[SalePrice]')    # y軸のラベル
plt.grid()   # グリッド線を表示

plt.show()   # 図の表示
```

### 【考察】

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


# データを学習用と検証用に分割
X_train, X_test, Y_train, Y_test = train_test_split(X_house_var, y_house_var, train_size = 0.7, test_size = 0.3, random_state = 0)

# 学習データを標準化
# scl = StandardScaler()
# scl.fit(X_train)
# X_train = scl.transform(X_train)
# X_test = scl.transform(X_test)



# 線形回帰による学習（fit）を行う。
lr = LinearRegression() # 線形回帰モデルのインスタンス化
lr.fit(X_train, Y_train) # 線形モデルの重みを学習
```

```python
from sklearn.metrics import mean_squared_error

Y_train_pred = lr.predict(X_train) # 学習データに対する目的変数を推定（predict）
Y_test_pred = lr.predict(X_test) # 検証データに対する目的変数を推定（predict）

# 評価（MSE）を行う。
print('MSE train data: ', mean_squared_error(Y_train, Y_train_pred)) # 学習データを用いたときの平均二乗誤差を出力
print('MSE test data: ', mean_squared_error(Y_test, Y_test_pred))         # 検証データを用いたときの平均二乗誤差を出力
```

### SVM

```python
from sklearn.svm import SVR

# 抽出
X_house_var = house_variables[['GrLivArea', 'YearBuilt']].values
y_house_var = house_variables[['SalePrice']].values

# 分割
X_train, X_test, Y_train, Y_test = train_test_split(
    X_house_var, y_house_var, train_size = 0.7, test_size = 0.3, random_state = 1)

# yの一次元化
Y_train = np.reshape(Y_train,(-1))
Y_test = np.reshape(Y_test,(-1))

# 学習
sv_reg = SVR()
sv_reg.fit(X_train, Y_train) 

# 推定
Y_train_pred_svr = sv_reg.predict(X_train)
Y_test_pred_svr = sv_reg.predict(X_test)

# 評価
svr_train_mse = mean_squared_error(Y_train, Y_train_pred_svr)
svr_test_mse = mean_squared_error(Y_test, Y_test_pred_svr)

print('MSE train data: ',svr_train_mse)
print('MSE test data: ',svr_test_mse)  
```

```python
# 可視化

# 単回帰分析でのGrLivAreaの散布図化
X_GrLivArea = house_variables[['GrLivArea']]

# データを学習用と検証用に分割
X_train, X_test, Y_train, Y_test = train_test_split(
    X_GrLivArea, y_house_var, train_size = 0.7, test_size = 0.3, random_state = 1)

# yの一次元化
Y_train = np.reshape(Y_train,(-1))
Y_test = np.reshape(Y_test,(-1))

# 学習
sv_reg = SVR(kernel='linear')
sv_reg.fit(X_train, Y_train) 

plt.figure(figsize=(12, 8))
plt.scatter(X_test, Y_test, color = 'blue')         # 説明変数と目的変数のデータ点の散布図をプロット
plt.plot(X_test, sv_reg.predict(X_test), color = 'red') # 回帰直線をプロット

plt.title('GrLivArea vs SalePrice with Regression Line')               # 図のタイトル
plt.xlabel('Above grade (ground) living area square feet [GrLivArea]') # x軸のラベル
plt.ylabel('[SalePrice]')    # y軸のラベル
plt.grid()                                 # グリッド線を表示

plt.show()                                 # 図の表示
```

```python
# 単回帰分析でのGrLivAreaの散布図化
X_YearBuilt = house_variables[['YearBuilt']]

# データを学習用と検証用に分割
X_train, X_test, Y_train, Y_test = train_test_split(
    X_YearBuilt, y_house_var, train_size = 0.7, test_size = 0.3, random_state = 1)

# yの一次元化
Y_train = np.reshape(Y_train,(-1))
Y_test = np.reshape(Y_test,(-1))

# 学習
sv_reg = SVR(kernel='linear')
sv_reg.fit(X_train, Y_train) 

plt.figure(figsize=(12, 8))
plt.scatter(X_test, Y_test, color = 'olive')         # 説明変数と目的変数のデータ点の散布図をプロット
plt.plot(X_test, sv_reg.predict(X_test), color = 'green') # 回帰直線をプロット

plt.title('GrLivArea vs SalePrice with Regression Line')               # 図のタイトル
plt.xlabel('[YearBuilt]') # x軸のラベル
plt.ylabel('[SalePrice]')    # y軸のラベル
plt.grid()                                 # グリッド線を表示

plt.show()                                 # 図の表示
```

### 決定木

```python
from sklearn.tree import DecisionTreeRegressor

# 抽出
X_house_var = house_variables[['GrLivArea', 'YearBuilt']].values
y_house_var = house_variables[['SalePrice']].values

# 分割
X_train, X_test, Y_train, Y_test = train_test_split(
    X_house_var, y_house_var, train_size = 0.7, test_size = 0.3, random_state=1)

# yの一次元化
Y_train = np.reshape(Y_train,(-1))
Y_test = np.reshape(Y_test,(-1))

# 学習
tree_reg = DecisionTreeRegressor(max_depth=3, random_state=0)
tree_reg.fit(X_train, Y_train) 

# 推定
Y_train_pred_tree = tree_reg.predict(X_train)
Y_test_pred_tree = tree_reg.predict(X_test)

# 評価
tree_train_mse = mean_squared_error(Y_train, Y_train_pred_tree)
tree_test_mse = mean_squared_error(Y_test, Y_test_pred_tree)

print('MSE train data: ',tree_train_mse)
print('MSE test data: ',tree_test_mse)  
```

```python
# 可視化

# 単回帰分析でのGrLivAreaの散布図化
X_GrLivArea = house_variables[['GrLivArea']]

# データを学習用と検証用に分割
X_train, X_test, Y_train, Y_test = train_test_split(
    X_GrLivArea, y_house_var, train_size = 0.7, test_size = 0.3, random_state = 1)

# yの一次元化
Y_train = np.reshape(Y_train,(-1))
Y_test = np.reshape(Y_test,(-1))

# データを標準化
# scl = StandardScaler()
# scl.fit(X_train)
# x_train = scl.transform(X_train)
# x_test = scl.transform(X_test)

# 学習
tree_reg = DecisionTreeRegressor(max_depth=3, random_state=1)
tree_reg.fit(X_train, Y_train) 

plt.figure(figsize=(12, 8))
plt.scatter(X_test, Y_test, color = 'blue')         # 説明変数と目的変数のデータ点の散布図をプロット
plt.plot(X_test, tree_reg.predict(X_test), color = 'red', label='depth=3') # 回帰直線をプロット

plt.title('GrLivArea vs SalePrice with Regression Line')               # 図のタイトル
plt.xlabel('Above grade (ground) living area square feet [GrLivArea]') # x軸のラベル
plt.ylabel('[SalePrice]')    # y軸のラベル
plt.grid()                                 # グリッド線を表示

plt.show()                                 # 図の表示
```

```python
# 単回帰分析でのGrLivAreaの散布図化
X_YearBuilt = house_variables[['YearBuilt']]

# データを学習用と検証用に分割
X_train, X_test, Y_train, Y_test = train_test_split(
    X_YearBuilt, y_house_var, train_size = 0.7, test_size = 0.3, random_state = 1)

# yの一次元化
Y_train = np.reshape(Y_train,(-1))
Y_test = np.reshape(Y_test,(-1))

# 学習
tree_reg = DecisionTreeRegressor(max_depth=3, random_state=1)
tree_reg.fit(X_train, Y_train) 

plt.figure(figsize=(12, 8))
plt.scatter(X_test, Y_test, color = 'olive')         # 説明変数と目的変数のデータ点の散布図をプロット
plt.plot(X_test, tree_reg.predict(X_test), color = 'green') # 回帰直線をプロット

plt.title('GrLivArea vs SalePrice with Regression Line')               # 図のタイトル
plt.xlabel('[YearBuilt]') # x軸のラベル
plt.ylabel('[SalePrice]')    # y軸のラベル
plt.grid()                                 # グリッド線を表示

plt.show()                                 # 図の表示
```

### ランダムフォレスト

```python
from sklearn.ensemble import RandomForestRegressor

# 抽出
X_house_var = house_variables[['GrLivArea', 'YearBuilt']].values
y_house_var = house_variables[['SalePrice']].values

# 分割
X_train, X_test, Y_train, Y_test = train_test_split(
    X_house_var, y_house_var, train_size = 0.7, test_size = 0.3, random_state=1)

# yの一次元化
Y_train = np.reshape(Y_train,(-1))
Y_test = np.reshape(Y_test,(-1))

# 学習
forest_reg = RandomForestRegressor(max_depth=3, random_state=0)
forest_reg.fit(X_train, Y_train) 

# 推定
Y_train_pred_forest = forest_reg.predict(X_train)
Y_test_pred_forest = forest_reg.predict(X_test)

# 評価
forest_train_mse = mean_squared_error(Y_train, Y_train_pred_forest)
forest_test_mse = mean_squared_error(Y_test, Y_test_pred_forest)

print('MSE train data: ',forest_train_mse)
print('MSE test data: ',forest_test_mse)  
```

```python
# 可視化

# 単回帰分析でのGrLivAreaの散布図化
X_GrLivArea = house_variables[['GrLivArea']]

# データを学習用と検証用に分割
X_train, X_test, Y_train, Y_test = train_test_split(
    X_GrLivArea, y_house_var, train_size = 0.7, test_size = 0.3, random_state = 1)

# yの一次元化
Y_train = np.reshape(Y_train,(-1))
Y_test = np.reshape(Y_test,(-1))

# データを標準化
# scl = StandardScaler()
# scl.fit(X_train)
# x_train = scl.transform(X_train)
# x_test = scl.transform(X_test)

# 学習
forest_reg = RandomForestRegressor(max_depth=3, random_state=0)
forest_reg.fit(X_train, Y_train) 

plt.figure(figsize=(12, 8))
plt.scatter(X_test, Y_test, color = 'blue')         # 説明変数と目的変数のデータ点の散布図をプロット
plt.plot(X_test, forest_reg.predict(X_test), color = 'red', label='depth=3') # 回帰直線をプロット

plt.title('GrLivArea vs SalePrice with Regression Line')               # 図のタイトル
plt.xlabel('Above grade (ground) living area square feet [GrLivArea]') # x軸のラベル
plt.ylabel('[SalePrice]')    # y軸のラベル
plt.grid()                                 # グリッド線を表示

plt.show()                                 # 図の表示
```

```python
# 単回帰分析でのGrLivAreaの散布図化
X_YearBuilt = house_variables[['YearBuilt']]

# データを学習用と検証用に分割
X_train, X_test, Y_train, Y_test = train_test_split(
    X_YearBuilt, y_house_var, train_size = 0.7, test_size = 0.3, random_state = 1)

# yの一次元化
Y_train = np.reshape(Y_train,(-1))
Y_test = np.reshape(Y_test,(-1))

# 学習
forest_reg = RandomForestRegressor(max_depth=3, random_state=0)
forest_reg.fit(X_train, Y_train) 

plt.figure(figsize=(12, 8))
plt.scatter(X_test, Y_test, color = 'olive')         # 説明変数と目的変数のデータ点の散布図をプロット
plt.plot(X_test, forest_reg.predict(X_test), color = 'green') # 回帰直線をプロット

plt.title('GrLivArea vs SalePrice with Regression Line')               # 図のタイトル
plt.xlabel('[YearBuilt]') # x軸のラベル
plt.ylabel('[SalePrice]')    # y軸のラベル
plt.grid()                                 # グリッド線を表示

plt.show()                                 # 図の表示
```
